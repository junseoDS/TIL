```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('ml-latest-small/movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
# 1. 사용자가 평가한 모든 영화의 전체 평균 평점

ratings['rating'].mean()
```

```python
# 2. 각 사용자별 평균 평점

ratings.groupby('userId')['rating'].mean()
```

```python
# 3. 각 영화별 평균 평점

pd.merge(movies,ratings,on='movieId',how='left',).groupby('movieId').mean()
```

```python
# 4. 평균 평점이 가장 높은 영화의 제목
# (동률이 있을 경우 모두 출력)

mr = pd.merge(movies,ratings,on='movieId',how='left',).groupby('movieId').mean()

#mr.sort_values(by=['rating'],ascending=False)
```

```python
mr[mr['rating']==mr['rating'].max()]
```

```python
# 5. Comedy영화 중 가장 평점이 낮은 영화의 제목을 출력하세요.
#   단, 동률이 있을 경우 모두 출력하고 title을 기준으로 오름차순 정렬하세요.
```

```python
mr = pd.merge(movies,ratings,on='movieId',how='outer')
mr[(mr['genres']=='Comedy')&(mr['rating']==mr['rating'].min())]
```
